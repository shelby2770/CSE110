```python
def even_check(str_1):
    dict_1= {}
    for i in str_1:
        if int(i) not in dict_1:
            if int(i)%2== 0:
                dict_1[int(i)]= "even"
            else:
                dict_1[int(i)]= "not even"
    return dict_1

def prime_check(str_1):
    dict_1= {}
    for i in str_1:
        if int(i) not in dict_1:
            counter= 0
            for j in range(1,int(i)+1):  
                if int(i)%j== 0: 
                    counter+=1
            if counter!= 2:
                dict_1[int(i)]= "not prime"
            else:
                dict_1[int(i)]= "prime"   
    return dict_1

def perfect_check(str_1):
    dict_1= {}
    for i in str_1:
        if int(i) not in dict_1:
            sum= 0
            for j in range(1,int(i)):
                if int(i)%j== 0:
                    sum+= j
            if sum==int(i):
                dict_1[int(i)]= "perfect"
            else:
                dict_1[int(i)]= "not perfect"
            sum= 0    
    return dict_1

def function_name():
    even= even_check(str_1)
    prime= prime_check(str_1)
    perfect= perfect_check(str_1)
    result= {}
    final_result= {}
    for k,v in even.items():
        result[k]= [v]
    for k,v in prime.items():
        if k in result:
            result[k].append(v)
    for k,v in perfect.items():
        if k in result:
            result[k].append(v)
    for k,v in result.items():
        final_result[k]= tuple(v)
    return final_result

str_1= input()
even_check(str_1)
prime_check(str_1)
perfect_check(str_1)
print(function_name())
```
